## Bike sharing: Predicción de demanda horaria

Se utilizarán los datos del concurso de Kaggle en el cual hay que predecir la demanda de bicicletas sobre la red Capital Bikeshare de Washington DC.

La función para evaluar el resultado es la siguiente:

$E_{bikes}(y,\hat{y})= \frac{1}{n}\sum_i(ln(y_i+1)-ln(\hat{y}_i+1))^2$

Donde $y, \hat{y} \in \rm I\!R^n$ son los vectores de observaciones y predicciones respectivamente.

### a) Carga y descripción de datos

```python
import pandas as pd
import numpy as np

dftrain = pd.read_csv('bike_sharing_train.csv')
dfval = pd.read_csv('bike_sharing_val.csv')
dftest = pd.read_csv('bike_sharing_test.csv')
ntrain = len(dftrain)
nval = len(dftrain) + len(dfval)
ntest = len(dftest)
df = pd.concat([dftrain,dfval,dftest])
print('\nSummary - dataframe completo:\n')
print df.describe()
df['hour'] = pd.to_datetime(df['datetime']).apply(lambda x: x.strftime('%H'))
df['hour'] = pd.to_numeric(df['hour'])

print "\nTamaño set de entrenamiento: ", ntrain
print "Tamaño set de pruebas: ", ntest
print "Tamaño set de validación: ", nval-ntrain
```

### b) Árbol de regresión base

Este modelo considerará como predictoras las variables *season, holiday, workingday, weather, temp, atemp, humidity, windspeed y hour*, la última fue agregada en el código y es un número de 0 a 23 que representa la hora del día. La variable *datetime* se ve reflejada en *hour*, y aunque deja de lado la tendencia según el tiempo del año, el campo *season* ayuda a reflejarla.

Los campos *casual* y *registered* seran ignorados por ahora, y se considerará *count* como el target, que es la suma de los otros dos.

```python
from sklearn.tree import DecisionTreeRegressor as Tree
def eval_bikemodel(y_predict,y_true):
    diff = np.log(y_predict+1.0) - np.log(y_true+1.0)
    return np.sqrt(np.sum(np.square(diff))/len(y_predict))

Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour']]
Ydf=df.ix[:,'count']

X_train = Xdf[0:ntrain].values
X_val = Xdf[ntrain:nval].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:ntrain].values
Y_val = Ydf[ntrain:nval].values
Y_test = Ydf[nval:].values

model = Tree(random_state=0)
model.fit(X_train,Y_train)
score_test = model.score(X_test,Y_test)
print "SCORE TEST=%f"%score_test

Y_pred_train = model.predict(X_train)
Y_pred_val = model.predict(X_val)
Y_pred_test = model.predict(X_test)
kagg_train = eval_bikemodel(Y_pred_train,Y_train)
kagg_val = eval_bikemodel(Y_pred_val,Y_val)
kagg_test = eval_bikemodel(Y_pred_test,Y_test)
print "KAGG EVAL TRAIN =%f"%kagg_train
print "KAGG EVAL VAL =%f"%kagg_val
print "KAGG EVAL TEST =%f"%kagg_test

plt.plot(Y_test, Y_pred_test, '.')
plt.xlabel("Valor real")
plt.ylabel("Valor prediccion")
plt.title("Prediccion vs Real en set de pruebas")
plt.plot( [0,1000],[0,1000] )
plt.show()
```

En el gráfico de dispersión ideal, todos los valores estarían sobre la diagonal, indicando una predicción perfecta. En este caso se ve que hay cierta tendencia hacia la diagonal pero hay bastante dispersión, el puntaje del set de pruebas según la función definida al comienzo es de 0.574.

```python
plt.plot(Y_train, Y_pred_train, '.')
plt.xlabel("Valor real")
plt.ylabel("Valor prediccion")
plt.title("Prediccion vs Real en set de entrenamiento")
plt.plot( [0,1000],[0,1000] )
plt.show()
```

Como se observa en el gráfico que compara los valores en el set de entrenamiento (además de que el error en el set de entrenamiento es de sólo 0.028 según la función de evaluación), el modelo está sobreajustado, es por eso que es necesario limitar el árbol para evitar el sobreajuste.

### c) Árbol con limitación de profundidad

```python
kagg_validacion = []
for i in range(2,26):
    model = Tree(random_state=0, max_depth = i)
    model.fit(X_train,Y_train)
    kagg_validacion.append(eval_bikemodel(model.predict(X_val),Y_val))

plt.plot(range(2,26), kagg_validacion)
plt.ylabel("Validation error")
plt.xlabel("Max Depth")
plt.show()
```

```python
kagg_validacion = []
for i in range(1,31):
    model = Tree(random_state=0, min_samples_leaf = i)
    model.fit(X_train,Y_train)
    kagg_validacion.append(eval_bikemodel(model.predict(X_val),Y_val))

plt.plot(range(1,31), kagg_validacion)
plt.ylabel("Validation error")
plt.xlabel("Min samples per leaf node")
plt.show()
```

Se revisan los parámetros max_depth y min_samples_leaf, estos limitan la profundidad del árbol y la cantidad minima de datos que pueden quedar en una hoja del árbol. El error baja en al rededor de un 10% en los casos óptimos por separado, buscaremos una buena mezcla de ambos parámetros.

```python
kagg_validacion = []
depth_valor = []
leaf_valor = []
for i in range(5,16):
    for j in range(5,16):
        model = Tree(random_state=0, max_depth = i, min_samples_leaf = j)
        model.fit(X_train,Y_train)
        kagg_validacion.append(eval_bikemodel(model.predict(X_val),Y_val))
        depth_valor.append(i)
        leaf_valor.append(j)

plt.plot(range(0,121), kagg_validacion)
plt.ylabel("Validation error")
plt.xlabel("Indice")
plt.show()

print "El error es minimo con profundidad", depth_valor[kagg_validacion.index(min(kagg_validacion))],"y minimo de ejemplos por hoja",leaf_valor[kagg_validacion.index(min(kagg_validacion))]
```

```python
model = Tree(random_state=0, max_depth = depth_valor[kagg_validacion.index(min(kagg_validacion))], min_samples_leaf = leaf_valor[kagg_validacion.index(min(kagg_validacion))])
model.fit(X_train,Y_train)
print "Error KAGG Val", eval_bikemodel(model.predict(X_val),Y_val)
print "Error KAGG Train", eval_bikemodel(model.predict(X_train),Y_train)
print "Error KAGG Test", eval_bikemodel(model.predict(X_test),Y_test)

plt.plot(Y_test, model.predict(X_test), '.')
plt.xlabel("Valor real")
plt.ylabel("Valor prediccion")
plt.title("Prediccion vs Real en set de entrenamiento")
plt.plot( [0,1000],[0,1000] )
plt.show()
```

Se genera el árbol que minimiza el error en el set de validación, éste tiene un error de pruebas de 0.49, aproximadamente 8 puntos mejor que el árbol por default. Además el error de entrenamiento ahora es 0.39, por lo que el sobreajuste no es tan grave.

### d) Agregando más atributos

Sólo se está considerando la hora del día del campo *datetime*, el modelo puede mejorar agregando más atributos como el día de la semana o el mes. Además se pueden aplicar transformaciones a las variables para que la predicción sea más fácil. Se probará predecir el logaritmo natural de la variable *count* + 1 (para evitar $ln(0)), esta es la transformación que se utiliza en la función de evaluación.

```python
df['cday'] = pd.to_datetime(df['datetime']).dt.dayofweek#0:lunes,6:domingo
df['cday'] = pd.to_numeric(df['cday'])
df['month'] = pd.to_datetime(df['datetime']).dt.month
df['month'] = pd.to_numeric(df['month'])
df['year'] = pd.to_datetime(df['datetime']).dt.year
df['year'] = pd.to_numeric(df['year'])

Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday','month','year']]
Ydf=np.log(df.ix[:,'count']+1)

X_train = Xdf[0:ntrain].values
X_val = Xdf[ntrain:nval].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:ntrain].values
Y_val = Ydf[ntrain:nval].values
Y_test = Ydf[nval:].values
```

Ejecutando las celdas anteriores, se determina que en el set de validación los mejores parámetros son depth = 16 y minimo de ejemplos = 13

```python
model = Tree(random_state=0, max_depth = 16, min_samples_leaf = 13)
model.fit(X_train,Y_train)
predict_val = np.exp(model.predict(X_val))-1
predict_train = np.exp(model.predict(X_train))-1
predict_test = np.exp(model.predict(X_test))-1
print "Error KAGG Val", eval_bikemodel(predict_val,np.exp(Y_val))
print "Error KAGG Train", eval_bikemodel(predict_train,np.exp(Y_train))
print "Error KAGG Test", eval_bikemodel(predict_test,np.exp(Y_test))

plt.plot(Y_test, model.predict(X_test), '.')
plt.xlabel("Valor real")
plt.ylabel("Valor prediccion")
plt.title("Prediccion vs Real en set de entrenamiento")
plt.plot( [0,10],[0,10] )
plt.show()
```

Con las nuevas variables el modelo llega a un error de 0.427 en el set de pruebas, que posicionaría al modelo cerca del top 500 del concurso pero sin romper esa barrera. 

### e) SVM no lineal

```python
Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday','month','year']]
Ydf=np.log(df.ix[:,'count']+1)

Xdf = pd.get_dummies(Xdf,columns=['season','weather','hour','cday','month','year'])

X_train = Xdf[0:ntrain].values
X_val = Xdf[ntrain:nval].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:ntrain].values
Y_val = Ydf[ntrain:nval].values
Y_test = Ydf[nval:].values

from sklearn.preprocessing import StandardScaler
scalerX = StandardScaler()
X_train = scalerX.fit_transform(X_train)
X_val = scalerX.fit_transform(X_val)
X_test = scalerX.fit_transform(X_test)
```

```python
from sklearn.svm import SVR
model = SVR()
model.fit(X_train,Y_train)

predict_val = np.exp(model.predict(X_val))-1
predict_train = np.exp(model.predict(X_train))-1
predict_test = np.exp(model.predict(X_test))-1
print "Error KAGG Val", eval_bikemodel(predict_val,np.exp(Y_val)-1)
print "Error KAGG Train", eval_bikemodel(predict_train,np.exp(Y_train)-1)
print "Error KAGG Test", eval_bikemodel(predict_test,np.exp(Y_test)-1)

plt.plot(Y_test, model.predict(X_test), '.')
plt.xlabel("Valor real")
plt.ylabel("Valor prediccion")
plt.title("Prediccion vs Real en set de entrenamiento")
plt.plot( [0,10],[0,10] )
plt.show()
```

Utilizando la SVR con los parámetros por default y aplicando la transformación de la predicción $ln(y+1)$ además de agregar los campos hora, día, mes y año, se logra una error en el set de pruebas de 0.326, que es superior al mejor resultado del desafío en Kaggle (aunque acá se está trabajando con un subset, por lo que podría no ser mejor que el modelo del top 1).

Ahora se puede además jugar con los parámetros del modelo para mejorarlo aún más. Se analizarán distintas combinaciones variando el kernel, el C y el epsilon de SVM.

```python
Cs = np.logspace(-3, 1,base=2, num=5)
kernels = ['poly','rbf']
epsilon = [0.01,0.05,0.1]

for k in kernels:
    print "========== Kernel",k,"==========" 
    for c in Cs:
        for ep in epsilon:
            model = SVR(C=c,epsilon=ep,kernel=k)
            model.fit(X_train,Y_train)
            predict_val = np.exp(model.predict(X_val))-1
            print "Error KAGG Val", eval_bikemodel(predict_val,np.exp(Y_val)-1), " C= ",c," e= ",ep
        print "------------------------------"
        

```

Se observa que el cálculo se demora más con un $C \geq 1$, ademásla calidad comienza a empeorar en el caso del kernel poly. El valor de epsilon crea una leve mejora al bajarlo a 0.05. 

La mejor opción encontrada es una SVR con kernel rbf y parámetros C=2.0 y e = 0.05, con un error de 0.288 en el set de validación. A continuación se muestra el comportamiento de éste en el set de pruebas.

```python
model = SVR(C=2,epsilon=0.05,kernel='rbf')
model.fit(X_train,Y_train)
predict_test = np.exp(model.predict(X_test))-1
print "Error KAGG Test", eval_bikemodel(predict_test,np.exp(Y_test)-1)

plt.plot(Y_test, model.predict(X_test), '.')
plt.xlabel("Valor real")
plt.ylabel("Valor prediccion")
plt.title("Prediccion vs Real en set de entrenamiento")
plt.plot( [0,10],[0,10] )
plt.show()
```

Finalmente, la máquina que obtiene los mejores resultados en el set de validación mejora levemente el error en el test set, llegando a 0.324.

### g) cross validation

```python
from sklearn import cross_validation

Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday','month','year']]
Ydf=np.log(df.ix[:,'count']+1)

X_train = Xdf[0:nval].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:nval].values
Y_test = Ydf[nval:].values

k_fold = cross_validation.KFold(len(X_train),10)
best_score = 100
best_leaf = 1
best_depth = 1
for leaf in range(1,9):
    for depth in range(8,18):
        score_cv = 0
        for k, (train, val) in enumerate(k_fold):
            model = Tree(random_state=0,min_samples_leaf=leaf, max_depth=depth)
            model.fit(X_train[train], Y_train[train])
            Ypred_val = np.exp(model.predict(X_train[val]))-1
            Ytrue_val = np.exp(Y_train[val])-1
            score_fold = eval_bikemodel(Ypred_val,Ytrue_val)
            score_cv += score_fold
        score_cv = score_cv / 10
        if score_cv < best_score:
            best_score = score_cv
            best_leaf = leaf
            best_depth = depth

#mejor encontrado          
model = Tree(random_state=0,min_samples_leaf=best_leaf, max_depth=best_depth)
model.fit(X_train,Y_train)
predict_test = np.exp(model.predict(X_test))-1
print "Mejor Error KAGG Test", eval_bikemodel(predict_test,np.exp(Y_test)-1), "Max Depth =", best_depth, "y min leaves per node = ", best_leaf

plt.plot(Y_test, model.predict(X_test), '.')
plt.xlabel("Valor real")
plt.ylabel("Valor prediccion")
plt.title("Prediccion vs Real en set de entrenamiento")
plt.plot( [0,10],[0,10] )
plt.show()
            
```

Para el árbol de decisión el resultado mejora al entrenarlo con el set de validación, la mejora es leve pero todo cuenta.

```python
Cs = np.linspace(0.5, 2, num = 7)

best_score = 100
best_c = 10

Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday','month','year']]
Xdf = pd.get_dummies(Xdf,columns=['season','weather','hour','cday','month','year'])
Ydf=np.log(df.ix[:,'count']+1)

X_train = Xdf[0:nval].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:nval].values
Y_test = Ydf[nval:].values

scalerX = StandardScaler()
X_train = scalerX.fit_transform(X_train)
X_test = scalerX.fit_transform(X_test)


for c in Cs:
    score_cv = 0
    for k, (train, val) in enumerate(k_fold):
        model = SVR(C=c,epsilon=0.05,kernel='rbf')
        model.fit(X_train[train], Y_train[train])
        Ypred_val = np.exp(model.predict(X_train[val]))-1
        Ytrue_val = np.exp(Y_train[val])-1
        score_fold = eval_bikemodel(Ypred_val,Ytrue_val)
        score_cv += score_fold
    score_cv = score_cv / 10
    if score_cv < best_score:
        print "Mejora con C=",c,"score = ",score_cv
        best_score = score_cv
        best_c = c

            
#mejor encontrado          
model = SVR(C=best_c,epsilon=0.05,kernel='rbf')
model.fit(X_train,Y_train)
predict_test = np.exp(model.predict(X_test))-1
print "Mejor Error KAGG Test", eval_bikemodel(predict_test,np.exp(Y_test)-1), "C =", best_c

plt.plot(Y_test, model.predict(X_test), '.')
plt.xlabel("Valor real")
plt.ylabel("Valor prediccion")
plt.title("Prediccion vs Real en set de entrenamiento")
plt.plot( [0,10],[0,10] )
plt.show()
```

Finalmente con cross-validation mejora un poco el error con la SVR, se tomó el kernel rbf y el epsilon como 0.05, el mejor C encontrado fue 1.5 y el error llega a 0.321, centésimas mejor que el error sin cross-validation.

### h) Ensamblado

Se predecirán por separado la cantidad de usuarios registrados y no registrados que arriendan bicicletas, ya que estas cantidades tienen patrones distintos se debería mejorar la calidad de la predicción.


```python
Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday','month','year']]
Xdf = pd.get_dummies(Xdf,columns=['season','weather','hour','cday','month','year'])
Ydf=np.log(df.ix[:,'count']+1)
Ydf_r=np.log(df.ix[:,'registered']+1)
Ydf_c=np.log(df.ix[:,'casual']+1)

X_train = Xdf[0:nval].values
X_test = Xdf[nval:].values
Y_train_c = Ydf_c[0:nval].values
Y_test_c = Ydf_c[nval:].values
Y_train_r = Ydf_r[0:nval].values
Y_test_r = Ydf_r[nval:].values
Y_test = Ydf[nval:].values

scalerX = StandardScaler()
X_train = scalerX.fit_transform(X_train)
X_test = scalerX.fit_transform(X_test)

model1 = SVR(C=1.5,epsilon=0.05,kernel='rbf')
model1.fit(X_train,Y_train_c)
predict_test_c = np.exp(model1.predict(X_test))-1

model2 = SVR(C=1.5,epsilon=0.05,kernel='rbf')
model2.fit(X_train,Y_train_r)
predict_test_r = np.exp(model2.predict(X_test))-1

predict_test = predict_test_r + predict_test_c
print "Error KAGG Test", eval_bikemodel(predict_test,np.exp(Y_test)-1)
```
